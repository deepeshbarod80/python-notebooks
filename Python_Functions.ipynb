# **Functions in Python**
### Purpose:

*   Perform a task
*   Return a value

```python
# Creating a simple function.
def greet():
  print("Hi There")
  print("This is Deepesh")

greet()
```

```python
def greet(first_name, last_name):
  print(f"Hi {first_name} {last_name}")
  print("Welcome to Abroad")

greet("Mosh", "Hamedani")
```

```python
# Function with return statement

def get_greeting(name):
  return(f"Hi {name}")

message = get_greeting("Deepesh")

print(message)
```

```python
# Function with return statement and writing inside the file

def get_greeting(name):
  return f"Hi {name}"

message = get_greeting("Deepesh")

file = open("content.txt", "w")
file.write(message)
```

# Keyword Arguments

```python
# A Sample format for keyword Arguments
def incriment(number, by):
  return number + by

print(incriment(2, by=1))
print('--------------------')

# Giving Explicit value of 'by'
def incriment(number, by=1):
  return number + by

print(incriment(2))
print('--------------------')

# Keyword argument with Addition operator
def incriment(number, by=5):
  return number + by

print(incriment(3))
print('----- Addition ---------------')

# Keyword argument with Substraction operator
def incriment(number, by=5):
  return number - by

print(incriment(15))
print('----- Substraction ---------------')

# Keyword argument with Multiplication operator
def incriment(number, by=5):
  return number * by

print(incriment(5))
print('----- Multiplication ---------------')

# Keyword argument with Division operator
def incriment(number, by=5):
  return number / by

print(incriment(20))
print('----- Division ---------------')

# Keyword argument with Modulus operator
def incriment(number, by=5):
  return number % by

print(incriment(22))
print('----- Modulus ---------------')
```

```python
# X Arguments = xargs

all_numbers = [2, 4, 6, 8]

def multiply(*number):
  total1 = 1
  for number in all_numbers:
    total1 *= number
  return total1

multiply()
```

```python
# Multiple Keyword arguments
# like **user

def save_user(**user):
  print(user)

save_user(id=204, age=32, Highiet=165)
```

```python
def save_user(**user):
  print(user[("id")])

save_user(id=204, age=32, Highiet=165)
```

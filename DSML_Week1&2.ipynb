```python
# Operators
# Arithmatic Operators

a = 10
b = 3
print(a + b)  # Addition operator.
print(a - b)  # Substraction operator.
print(a * b)  # Multiplication operator.
print(a / b)  # Division operator.
print(a // b) # Floor Division operator. Note: "We consider only lowest value so values are ignored after dot.
print(a % b)  # Modulus Operator. Note: 10/3 = 3 times devided and 1 is remaining.
print(a ** b) # Exponenetial Operator. Note: multiply 10 by 3 times means 1000 is result.
```

```python
# Comparison Operators

a = 10
b = 20
print(a == b) # Equal Operator. False
print(a != b) # Not-equal Operator. True
print(a > b)  # Greater than Operator. False
print(a < b)  # Less than Operator. True
print(a <= b) # Less than or Equal Operator. True
print(a >= b) # Greater than or Equal Operator. False
```

```python
# Logical Operators
# Logical 'and' Operator
a = 5
b = 2
c = 3
d = 200

print(a>b and a>c)  # True and True = True
print(a>b and a<c)  # True and False = False
print(a<b and a>c)  # False and True = False
print(a<b and a<c)  # False and False = False
print(a>b and c)    # True and 3 = 3
print(a>b and c and d) # True and 3 and 200 = 200
print(a<b and c)    # False and 3 = False
print(a<b and c and d) # False and 3 and 200 = False
```

```python
# Logical 'Or' Operator
a = 5
b = 2
c = 3
d = 200

print(a>b or a>c)  # True or True = True
print(a>b or a<c)  # True or False = True
print(a<b or a>c)  # False or True = True
print(a<b or a<c)  # False or False = False
print(a>b or c)    # True and 3 = True
print(a>b or c or d) # True and 3 and 200 = True
print(a<b or c)    # False and 3 = 3
print(a<b or c or d) # False and 3 and 200 = 3
```

```python
# Bitwise Operators

# Binary numbers: 0,1    10 = 1010

a = 10 # 1 0 1 0 (Binary conversion of 10)
b =  4 # 0 1 0 0 (Binary conversion of 4)

# bin(10) # If run this you will get result in binary number that is '1010'
# print(0b1110) # Run print(0b.....(binary digits)) to a Numeric Digit.
# bin(4) # If run this you will get result in binary number that is '0100'
# print(0b0010) # Run print(0b.....(binary digits)) to a Numeric Digit



# 1. 'And' Operator: '&'
# '&' compares each bit of 1st operand with the 2nd.
# If both bit are 1, the corresponding bit is set to 1 otherwise 0

# 10 = 1 0 1 0
# 4  = 0 1 0 0
# ------------
#      0 0 0 0 = 0

result = a & b
print(result)
# -------------------------------------------------------------------

# 2. 'Or' Operator: '|'
# '|' compares each bit of the 1st operand with the 2nd.
# If either bit is 1, the corresponding result bit is set 1, otherwise the result bit is set to 0.
result = a | b
print(result)
# 10 = 1 0 1 0
#  4 = 0 1 0 0
# ------------
#      1 1 1 0 = 14
# -------------------------------------------------------------------

# 3. Not Operator
# '~' compares True and False values and give opposite resuls


# 4. 'XOR' Operator: '|'
# '^' compares each bit of the 1st operand with the 2nd.
# If one bit is 1 (but not both), the corresponding result bit is set 1, otherwise the result bit is set to 0.
result = a ^ b
print(result)
# 10 = 1 0 1 0
#  4 = 0 1 0 0
# ------------
#      1 1 1 0 = 14
# -------------------------------------------------------------------

# 5. Left Shift operator
a = 10 # bn(10) = 1010
result = a << 2 # a<<2 means 00 will increase in the end.
# means '1010' will turn into '101000' that value is 40 in numeric.
print(result)

# 6. Right Shift operator
a = 10 # bn(10) = 1010
result = a >> 2 # a<<2 means 00 will increase in the starting.
# means '1010' will turn into '001010' that value is 40 in numeric.
print(result)
```

```python
 # Nested Loops

 # if We want result like the following
 # *       i = 0, then j in range(1)
 # * *     i = 1, then j in range(2)
 # * * *   i = 2, then j in range(3)
 # * * * * i = 3, then j in range(4)

rows = 4

for i in range(rows):
  for j in range(i+1):
    print("*", end = " ")
  print() # \n new line charactor
```

```python
# While

count = 1
while count <= 5: # while will run untill value of count of 5, if value is 6 it will not run.
  print(count)    # printing value of count
  count += 1      # count = count + 1

# 1st Interaction
# Step 1: print value of count 1
# Step 2: count will be incremented by 1 count = count + 1
# ---------------------------------------------------------------

# Now Updated count = 3
# while count <= 5: still condition is true or holds

# 2nd Interaction
# Step 1: print value of count 1
# Step 2: count will be incremented by 1 count = count + 1
#----------------------------------------------------------------

# Now Updated count = 6
# while count <= 5: Now condition is false now
```

```python
# 'for' we use generally we know when to stop
# 'While' we use generally when stop conditions will be satisfied, we don't know
# We have a list names = [Alice, Tom, Marry, James]
# Salary = [30000, 40000, 45000, 40000]
# give 5% bonus to James as he is performing well.

names = ["Alice", "Tom", "Marry", "James"]
salary = [30000, 40000, 45000, 40000]

print("old salary is :")

# First sub-problem is we will iterate both lists
# We need to update salary of Jaames only

for i in range(len(names)):
  if names[i] == "James":
    increment = salary[i]*0.05
    salary[i] += increment

print("updated salary is:")
print(salary)
```

```python
# here we are incresing 20% salary of Alice
for i in range(len(names)):
  if names[i] == "Alice":
    increment = salary[i]*0.20
    salary[i] += increment
print("updated salary is:")
print(salary)
```
